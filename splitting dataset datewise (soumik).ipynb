```python
previous_line = None
records=[]

with open("Aruba_without_trans.txt", 'r') as txt:
    for line in txt:
        parts = line.strip().split()
        if len(parts) < 6:
            for i in range(len(parts), 6):
                parts.append(previous_line[i])
        # # Discard the second elements that do not start with 'M'
        # if not parts[2].startswith('M'):
        #     continue
        records.append(parts)
        previous_line= parts
```

```python
df= pd.DataFrame(records, columns=['Date','Time','Sensor','Status','Activity','Label'])
```

```python
df
```

```python
# Get the date range from the user
start_date = input("Enter the start date (dd-mm-yyyy): ")
end_date = input("Enter the end date (dd-mm-yyyy): ")
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)  # assuming your date format is "%d-%m-%Y"
start_date = pd.to_datetime(start_date, dayfirst=True)
end_date = pd.to_datetime(end_date, dayfirst=True)

new_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)].copy()
```
